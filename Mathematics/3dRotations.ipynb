```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from ipywidgets import interact, interactive, fixed, widgets
plt.rcParams['font.family']='serif'
plt.rcParams['font.weight']='light'
plt.rcParams['font.size']=14
figsize = (12,6)
dataDir = '/Users/sdporzio/Data/'
```

```python
from numpy import cos
from numpy import sin
from numpy import sqrt

def rotmat(theta,phi,i,j):
    rotmat = [
        [cos(phi),sin(phi),0],
        [-cos(theta)*sin(phi),cos(theta)*cos(phi),-sin(theta)],
        [-sin(theta)*sin(phi),cos(phi)*sin(theta),cos(theta)]
    ]
#     rotmat = [
#         [cos(phi),-cos(theta)*sin(phi),-sin(theta)*sin(phi)],
#         [sin(phi),cos(theta)*cos(phi),cos(phi)*sin(theta)],
#         [0,-sin(theta),cos(theta)]
#     ]
    return rotmat[i][j]

v = [0.6,0,0.5]
r = sqrt(pow(v[0],2) + pow(v[1],2) + pow(v[2],2))
stheta = np.rad2deg(np.arccos(v[2]/float(r)))
if v[0]==0: sphi = 180
else: sphi = np.rad2deg(np.arctan(v[1]/float(v[0])))
print stheta, sphi

def look(elevation=10,azimuth=75,theta=stheta,phi=sphi):
    theta = np.deg2rad(theta)
    phi = np.deg2rad(phi)
    fig = plt.figure(figsize=figsize)
    ax = fig.add_subplot(111, projection='3d')
    ax.plot([0,1],[0,0],[0,0],color='black',lw=3)
    ax.plot([0,0],[0,1],[0,0],color='black',lw=3)
    ax.plot([0,0],[0,0],[0,1],color='black',lw=3)
    ax.plot([0,rotmat(theta,phi,0,0)],[0,rotmat(theta,phi,0,1)],[0,rotmat(theta,phi,0,2)],color='g',lw=3)
    ax.plot([0,rotmat(theta,phi,1,0)],[0,rotmat(theta,phi,1,1)],[0,rotmat(theta,phi,1,2)],color='g',lw=3)
    ax.plot([0,rotmat(theta,phi,2,0)],[0,rotmat(theta,phi,2,1)],[0,rotmat(theta,phi,2,2)],color='g',lw=3)
    ax.plot([0,v[0]],[0,v[1]],[0,v[2]],color='r',lw=5)
    ax.view_init(elev=elevation, azim=azimuth)
    
w_elev = (0,90,5)
w_azim = (0,90,5)
w_theta = (-180,180,5)
w_phi = (-180,180,5)

interact(look, elevation=w_elev,azimuth=w_azim,theta=w_theta,phi=w_phi)
```
