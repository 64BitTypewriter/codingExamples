```python
import ROOT
from IPython.display import Image as Plot
ROOT.gROOT.SetBatch(1)
```

```python
h = ROOT.TH1F("gauss","Example histogram",100,-4,4)
h.FillRandom("gaus")
```

```python
c = ROOT.TCanvas("myCanvasName","The Canvas Title",1000,500)
h.Draw()
```

```python
c.SaveAs("tree.png")
Plot("tree.png")
```
