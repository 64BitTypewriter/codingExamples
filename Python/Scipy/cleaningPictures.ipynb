```python
%matplotlib inline
import matplotlib.pyplot as plt
import pylab as pl
import scipy.fftpack as fourier
```

```python
moonland = pl.imread('moonlanding.png')
plt.imshow(moonland,cmap='Greys_r')
```

```python
sigFreq = fourier.fft2(moonland)
plt.figure(figsize=(15,5))
plt.plot(abs(sigFreq[0]))
plt.show()

def customPeakFinder(array,threshold):
    indColl = []
    for i,val in enumerate(array):
        if val > threshold:
            indColl.append(i)
    return indColl

peakInds = customPeakFinder(abs(sigFreq[0]),30000)

# for i in range(len(sigFreq)):
#     for j in peakInds:
#         sigFreq[i][j] = 0
        
plt.figure(figsize=(15,5))
plt.plot(abs(sigFreq[0]))
plt.show()
```
