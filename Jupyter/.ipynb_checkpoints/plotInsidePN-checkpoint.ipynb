## Example code
Plotting a TCanvas object in browser and example formulas: $a = \sqrt{c+d} - \log{ef}$

Importing useful stuff (with nice python TObject destroying function included)

```python
import ROOT
from IPython.display import Image as Plot
ROOT.gROOT.SetBatch(1)

def destruct(object_):
    if issubclass(type(object_), ROOT.TObject):
        object_.IsA().Destructor(object_)
```

Creating histogram and filling hit

```python
h = ROOT.TH1F("gauss","Gaussian",100,-4,4)
h2 = ROOT.TH1F("exp","Exponential",100,0,5)
h.FillRandom("gaus",10000)
h2.FillRandom("expo",10000)
```

All the graph personalization should go in one block

```python
c = ROOT.TCanvas("myCanvasName","The Canvas Title",1000,500)
c.Divide(2,1)
c.cd(1)
h.SetLineColor(ROOT.kBlue)
h.Draw("EH")
c.cd(2)
h2.SetLineColor(ROOT.kRed)
h2.Draw("EH")
```

Finally this is plotting the histogram on the browser

```python
c.SaveAs("histo.png")
destruct(c)
destruct(h)
destruct(h2)

Plot("histo.png")
```
